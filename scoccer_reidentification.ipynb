```python
import cv2
from ultralytics import YOLO
from google.colab.patches import cv2_imshow
```

```python
MODEL_PATH = "/content/drive/MyDrive/best.pt"
VIDEO_PATH = "/content/drive/MyDrive/football.mp4"
OUTPUT_VIDEO_PATH = "/content/drive/MyDrive/football_output.mp4"
```

```python
model = YOLO(MODEL_PATH)
```

```python
cap = cv2.VideoCapture(VIDEO_PATH)
if not cap.isOpened():
    print("Error: Could not open video file.")
else:

    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    print(f"\nVideo Properties: {frame_width}x{frame_height} @ {fps} FPS, Total Frames: {total_frames}")

    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(OUTPUT_VIDEO_PATH, fourcc, fps, (frame_width, frame_height))

    print("\nStarting player tracking process...")
    frame_count = 0

    while cap.isOpened():

        success, frame = cap.read()

        if success:
            frame_count += 1
            print(f"Processing frame {frame_count}/{total_frames}...")

            results = model.track(frame, persist=True, tracker='bytetrack.yaml')

            annotated_frame = results[0].plot()

            out.write(annotated_frame)

            if frame_count in [1, 50, 100, 150, 200, 250, 300, 350, 400, 450]:
                 print(f"\n--- Displaying annotated frame {frame_count} ---")
                 cv2_imshow(cv2.resize(annotated_frame, (frame_width // 2, frame_height // 2))) # Resize for display

        else:
            # Break the loop if the end of the video is reached
            break

    print("\nProcessing complete. Releasing video resources.")
    cap.release()
    out.release()
    print(f"Successfully saved tracked video to: {OUTPUT_VIDEO_PATH}")
```
